```python
list_14 = []
list_13 = []
m_list = []
for dir_path, subdir_list, file_list in os.walk(path):
    for x in file_list:
        if '14' in x and x.endswith('.csv'):
            list_14.append(x)
        if '13' in x:
            list_13.append(x)
        if x.endswith('.m'):
            fpath = dir_path + '/' +  x
            stat = os.stat(fpath)
            created_at=  stat.st_birthtime
            data_dict = {'filename':x,'created_at':created_at}
            m_list.append(data_dict)
            
    
    
```
